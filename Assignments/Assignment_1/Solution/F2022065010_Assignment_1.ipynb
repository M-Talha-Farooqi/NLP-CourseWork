# Regex Expressions

# Part 1: Rgular Expression Extraction
We will extract & visualize existing regular expressions in our dataset.

```python
# Necessary Libraries
import re
```

### 1. Extracting Words as Tokens(\w+)

This regex collects sequences made of letters, numbers, or underscores. It groups characters into meaningful word-like chunks such as hello, abc_123, or identifiers used in code.

```python
df['Words_Tokens'] = df['Text'].str.findall(r'\w+')
df
```

### 2. Extracting Digits from Text (\d+)
- This regex finds numbers inside text. \d means any digit (0â€“9), and + means one or more. It helps extract years, prices, phone numbers, and counts.

```python
df['Digits'] = df['Text'].str.findall(r'\d+')
df
```

### 3. Extracting Hashtags from Text (#\w+)

Matches hashtags beginning with # followed by letters, numbers, or underscores. Used for Twitter/Instagram data, e.g., #TerrorAlert.

```python
df['Hashtags'] = df['Text'].str.findall(r'#\w+')
df
```

### 4. Extracting Mentions / Usernames (@\w+)

Finds social media mentions starting with @, such as @IntelWatch. Useful for scraping Twitter, comments, or tagging systems.

```python
df['Mentions'] = df['Text'].str.findall(r'@\w+')
df
```

### 5. Detecting Acronyms or Uppercase Words ([A-Z]{2,})

[A-Z] selects uppercase letters; {2,} means two or more in a row. Matches words like SOS, CPU, ALERT. Helpful for abbreviations and warnings.

```python
df['UPPER-Case/Acronym'] = df['Text'].str.findall(r'[A-Z]{2,}')
df
```

### 6. Extracting Capitalized Words

This regex identifies words that start with a capital letter followed by lowercase letters, such as names, places, or proper nouns. It helps pull out entities like â€œLahoreâ€, â€œIntelWatchâ€, â€œVisitâ€, or any other capitalized token that appears inside the text.

```python
df['Capitalized'] = df['Text'].str.findall(r'\b[A-Z][a-z]+\b')
df
```

## 7. Matching Exact Whole Words (\bword\b)

\b is a word boundary. This pattern matches the exact word word, not parts of bigger words like sword. Good when searching precise keywords.

```python
# Extracting Security and Terror-Related Keywords Using Regex

df['Keywords'] = df['Text'].str.findall(
    r'\b(terror|alert|terroralert|cctv|ip|logs?|reports?|secure|email|data|files?)\b',
    flags=re.IGNORECASE
)

df
```

### 8. Extracting Email Addresses

This pattern searches for email formats inside the text. It looks for a username (letters/digits/symbols), the @ symbol, a domain name, and a valid extension like .com or .pk. The regex returns all email addresses found in each row of the dataset.

```python
df['Emails'] = df['Text'].str.findall(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}')
df
```

### 9. Extracting URLs (HTTP/HTTPs/www) from Text

This regex finds web links inside the text. It matches both http and https URLs, as well as links starting with www.. It captures full website paths until a space appears, which helps detect complete URLs in sentences, tweets, logs, or messages.

```python
df['URLs'] = df['Text'].str.findall(r'(https?://\S+|www\.[^\s]+)')
df
```

### 10. Matching Any Single Character Using the Dot Operator

The dot . matches any one character except a newline.
So the pattern a.b will find three-letter sequences that start with a, end with b, and have any character in the middle â€” for example: aab, acb, a0b.

```python
# Dot (.) example: matches a.b â†’ aab, acb
df['Dot_a.b'] = df['Text'].str.findall(r'a.b')
df
```

### 11. Extracting IPv4 Addresses
Captures IPs like 192.168.0.1 present in logs.


```python
df['IPv4'] = df['Text'].str.findall(r'\b(?:\d{1,3}\.){3}\d{1,3}\b')
df
```

### 12. Extracting File Paths (Windows or Linux)
Finds Windows paths like C:\Security\logs\case_21.txt and Linux paths like /var/data/reports/case-21.md.

```python
df['File_Paths'] = df['Text'].str.findall(r'([A-Za-z]:\\[\w\\.-]+|\/[\w\/\.-]+)')
df
```

### 13. Extracting Timestamps (HH:MM AM/PM)
Detects times like 11:45 PM, 07:30 PM.

```python
df['Timestamps'] = df['Text'].str.findall(r'\b\d{1,2}:\d{2}\s?(?:AM|PM)\b', flags=0)
df
```

### 14. Extracting Phone Numbers (International-ish)
Matches patterns like +92-300-12345678 with optional spaces/dashes.

```python
df['Phone_Numbers'] = df['Text'].str.findall(r'\+?\d{1,3}[-\s]?\d{3}[-\s]?\d{6,}')
df
```

### 15. Extracting Naming Styles (camelCase, snake_case, kebab-case)
Useful for code/token style detection.

```python
df['CamelCase'] = df['Text'].str.findall(r'\b[a-z]+(?:[A-Z][a-z]+)+\b')
df['snake_case'] = df['Text'].str.findall(r'\b[a-z]+(?:_[a-z]+)+\b')
df['kebab-case'] = df['Text'].str.findall(r'\b[a-z]+(?:-[a-z]+)+\b')
df
```

### 16. Extracting Percentages
Finds percentages like 99%.

```python
df['Percentages'] = df['Text'].str.findall(r'\b\d+%')
df
```

### 17. Detecting Multiple Spaces or Tabs

This pattern finds unnecessary whitespace in the text. It matches places where two or more spaces appear, or where tabs are used. It helps identify formatting issues that need normalization.

```python
df['Extra_Spaces_Tabs'] = df['Text'].str.findall(r'(?:\s{2,}|\t+)')
df
```

# Part 2: Regex Cleaning
We will clean our DataSet with all above extracted regular expressions.

```python
# Creating new DataFrame
Clean_Data = df[['ID','Text']].copy()
```

```python
# Remove URLs
Clean_Data['Clean_Text'] = Clean_Data['Text'].str.replace(r'(https?://\S+|www\.[^\s]+)', '', regex=True)

# Remove Emails
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}', '', regex=True)

# Remove Mentions
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'@\w+', '', regex=True)

# Remove Hashtags
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'#\w+', '', regex=True)

# Remove IPv4
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\b(?:\d{1,3}\.){3}\d{1,3}\b', '', regex=True)

# Remove File Paths
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'([A-Za-z]:\\[\w\\.-]+|\/[\w\/\.-]+)', '', regex=True)

# Remove Timestamps
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\b\d{1,2}:\d{2}\s?(?:AM|PM)\b', '', regex=True)

# Remove Phones
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\+?\d{1,3}[-\s]?\d{3}[-\s]?\d{6,}', '', regex=True)

# Remove Percentages
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\b\d+%', '', regex=True)

# Remove a.b patterns
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'a.b', '', regex=True)

# Remove Numbers
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\d+', '', regex=True)

# Remove UPPER acronyms
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\b[A-Z]{2,}\b', '', regex=True)

# Remove extra spaces / tabs
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'(?:\s{2,}|\t+)', ' ', regex=True)

# Normalize spaces
Clean_Data['Clean_Text'] = Clean_Data['Clean_Text'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

# Pre Processing

## Step 1: Tokenization

```python
# Necessary Libraries
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')

Clean_df = Clean_Data[['ID', 'Clean_Text']].copy()
Clean_df['Tokens'] = Clean_df['Clean_Text'].apply(word_tokenize)
```

```python
Clean_df
```

## Step 2: Remove Punctuations

```python
import string

# function to remove punctuation tokens
def remove_punctuation(tokens):
    return [word for word in tokens if word not in string.punctuation]

# apply to Clean_df
Clean_df['CleanTokens'] = Clean_df['Tokens'].apply(remove_punctuation)

Clean_df
```

## Step 3: Removal of StopWords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# Stopword set
stop_words = set(stopwords.words('english'))

# Function to remove stopwords
def remove_stopwords(tokens):
    return [word for word in tokens if word.lower() not in stop_words]

# Apply on CleanTokens column
Clean_df['NoStopwords'] = Clean_df['CleanTokens'].apply(remove_stopwords)

Clean_df
```

## Step 4: Lower Casing

```python
# Define the function to convert tokens to lowercase
def lowercase_tokens(tokens):
    return [word.lower() for word in tokens]

# Apply the function to the 'CleanTokens' column
Clean_df['LowercaseTokens'] = Clean_df['NoStopwords'].apply(lowercase_tokens)

Clean_df
```

## Step 5: Applying Stemming

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

# Define the stemming function
def stem_text(tokens):
    return [stemmer.stem(word) for word in tokens]

# Apply to CleanTokens
Clean_df['StemTokens'] = Clean_df['LowercaseTokens'].apply(stem_text)

Clean_df
```

## Step 6: Lemmatization

```python
from nltk.stem import WordNetLemmatizer

# Download WordNet resources if not already installed
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Define the lemmatization function
def lemmatize_text(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]

# Apply on stemmed tokens
Clean_df['LemmaTokens'] = Clean_df['LowercaseTokens'].apply(lemmatize_text)

Clean_df
```

## Joining Tokens Back into Clean Text

```python
def join_tokens(tokens):
    return ' '.join(tokens)

Clean_df['clean_Lemma_text'] = Clean_df['LemmaTokens'].apply(join_tokens)

Clean_df
```
